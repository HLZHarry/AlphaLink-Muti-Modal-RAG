```python
from sec_edgar_downloader import Downloader
from google.colab import drive
import os
import time
```

```python
# 1. Mount Google Drive to save data permanently
drive.mount('/content/drive')
```

```python
# 2. Setup Project Directories
# This creates a folder structure in your Drive: AlphaLink-RAG/data/raw
BASE_DIR = "/content/drive/MyDrive/AlphaLink-RAG"
DATA_DIR = os.path.join(BASE_DIR, "data/raw")
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
# 3. Initialize Downloader
# IMPORTANT: SEC requires "Company Name Email" to avoid being blocked.
USER_AGENT = "PersonalProject (l52he@outlook.com)"
dl = Downloader("AlphaLinkProject", "l52he@outlook.com", DATA_DIR)
```

```python
# 4. Define Company Scopes
# US companies use 10-K/10-Q; International use 20-F/6-K
targets = {
    "NVDA": ["10-K", "10-Q"],
    "TSM":  ["20-F", "6-K"],
    "ASML": ["20-F", "6-K"]
}
```

```python
# 5. Execution Loop with Rate Limiting (The "Polite" Scraper)
START_DATE = "2023-01-01"

print(f"🚀 Starting download to: {DATA_DIR}")

for ticker, forms in targets.items():
    for form_type in forms:
        try:
            print(f"--- Fetching {form_type} for {ticker} ---")

            # download_details=True ensures we get the full filing content
            # limit=None pulls everything since the START_DATE
            dl.get(form_type, ticker, after=START_DATE, download_details=True)

            # Wait 1.5 seconds after every single filing to respect 10 req/sec limit
            # Colab is fast; the SEC is sensitive.
            time.sleep(1.5)

        except Exception as e:
            if "503" in str(e):
                print(f"⚠️ Rate limit hit. Cooling down for 30s...")
                time.sleep(30)
            else:
                print(f"❌ Error downloading {form_type} for {ticker}: {e}")

print("✅ Step 1 Complete: All Anchor documents are secured in Google Drive.")
```
